```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data[:100, [0,3]]  # we only take the first two features.
Y = iris.target
y=Y[:100]
y[:50]=-1
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.title('iris data set')
plt.show()
```

```python
X1=np.mean(X,axis=0)
s1=np.std(X,axis=0)
X=(X-X1)/s1
on=np.ones(100)


Xone=np.c_[on,X]


np.random.seed(0)
nn_input=3
nn_output=1
W=np.random.rand(nn_input)
epsilon=0.1
lossT=[]
d=np.dot(Xone,W)*y
d1=d[d<0]
yn=y[d<0]
Xn=Xone[d<0]
iteration=[]
losses=[]

for i in range (10):
    d=np.dot(Xone,W)*y
    d1=d[d<0]
    yn=y[d<0]
    Xn=Xone[d<0]
    dw=epsilon*np.dot(Xn.T,yn)
    W=W+dw
    Ep=np.dot(Xn,W)*yn
    loss=np.sum(Ep)
    lossT.append(loss)
    iteration.append(i)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
xmy=np.linspace(np.min(X[:,0]),np.max(X[:,0]))
my=-(W[0]+W[1]*xmy)/W[2]
plt.plot(xmy,my)
plt.xlim([np.min(X[:,0]),np.max(X[:,0])])

plt.show()   
print W

```
